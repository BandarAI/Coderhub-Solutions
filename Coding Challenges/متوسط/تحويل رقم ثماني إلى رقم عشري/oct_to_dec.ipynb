# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل متغير من نوع int يعبر عن قيمة ثمانية octal number، ثم تقوم الدالة بإرجاع النتيجة بعد التحويل الى قيمة عشرية decimal number بنوع int
</div>

# Solution

```python
from typing import List
def oct_to_dec(octal_number: int) -> int:
    return int(str(octal_number), 8)
```

```python
oct_to_dec(octal_number = 5252)
```

```python
oct_to_dec(octal_number = 1534)
```

```python
oct_to_dec(octal_number = 412)
```

```python
oct_to_dec(octal_number = 1000)
```
