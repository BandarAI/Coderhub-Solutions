# Problem Description

<div dir="rtl" style="text-align: right;">
مثاال:
قبل الفلترة :
Goodbye bye bye world world world

بعد الفلترة :
Goodbye bye world
</div>


# Solution

```python
from typing import List
def removeRepeatedWords(sentence: str) -> str:
    words = sentence.split()
    result = []

    for word in words:
        if not result or result[-1].lower() != word.lower():
            result.append(word)
    return ' '.join(result)
```

```python
removeRepeatedWords(sentence = 'Goodbye bye bye world world world')
```

```python
removeRepeatedWords(sentence = 'Sam went went to to to his business')
```

```python
removeRepeatedWords(sentence = 'Reya is is the the best player in eye eye game')
```

```python
removeRepeatedWords(sentence = 'in inthe')
```
