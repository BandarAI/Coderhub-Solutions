["غترة مقصب", "زبون", "بشت"]

✨ النتيجة:
أدرك سالم أنه يجب عليه ارتداء غترة مقصب، زبون، وبشت، وهو اللباس التقليدي لأهل المنطقة الوسطى.

ممكن انها تساعدك :
المنطقه الوسطى 

غتره ، مقصب ، زبون ، بشت 

المنطقة الشرقيه 

غتره ، مقصب ، ثوب مرودن  ، دقله 

المنطقة الغربية

طاقية ، عمامة ، ثوب ، شاية 

المنطقة الجنوبية  

عصابة ، بيدي ، ثوب 

المنطقة الشمالية 

عقال ، غتره ، بشت ، ثوب مرودن

## Solution

```python
from typing import List

def get_saudi_traditional_outfit(region: str) -> List[str]:
    region_dict = {
        "الوسطى": ["غترة مقصب", "زبون", "بشت"],
        "الشرقية": ["غترة مقصب", "ثوب", "مرودن", "دقلة"],
        "الغربية": ["طاقية", "عمامة", "ثوب", "شاية"],
        "الجنوبية": ["عصابة بيدي", "ثوب"],
        "الشمالية": ["عقال", "غترة", "بشت", "ثوب مرودن"],
    }
    
    return region_dict.get(region, [])
```

```python
get_saudi_traditional_outfit(region = 'الوسطى')
```

```python
get_saudi_traditional_outfit(region = 'الشرقية')
```

```python
get_saudi_traditional_outfit(region = 'الغربية')
```

```python
get_saudi_traditional_outfit(region = 'الجنوبية')
```
