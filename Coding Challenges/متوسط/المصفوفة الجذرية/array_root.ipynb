# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل مصفوفة من نوع double. تقوم الدالة بحساب الجذر التربيعي لكل عنصر في المصفوفة ثم تقوم بإرجاع النتيجة في مصفوفة جديدة.

</div>


# Solution

```python
from typing import List
def array_root(arr: List[float]) -> List[float]:
    for number in range(len(arr)):
        arr[number] = arr[number] ** 0.5
    return arr
```

```python
array_root(arr = [4, 81, 49])
```

```python
array_root(arr = [15, 6, 3])
```

```python
array_root(arr = [121, 144, 169])
```

```python
array_root(arr = [4.2, 6.9])
```

## Computational Complexity

### Time Complexity: $O(n)$
### Space Complexity: $O(1)$
