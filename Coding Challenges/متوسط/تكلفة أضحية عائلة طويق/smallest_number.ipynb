# Problem Description

<div dir="rtl" style="text-align: right;">
مهمتك هي كتابة دالة calculateSacrificeContributions التي توزع التكلفة على أفراد العائلة حسب حصصهم، وتعيد مصفوفة توضح ما يدفعه كل فرد (قيمة مقربة إلى منزلتين عشريتين).
</div>


# Solution

```python
from typing import List
def calculateSacrificeContributions(names: List[str], shares: List[float], totalCost: float) -> List[float]:
    return [share * totalCost for share in shares]
```

```python
calculateSacrificeContributions(names = ['أحمد', 'سارة'], shares = [0.5, 0.5], totalCost = 200)
```

```python
calculateSacrificeContributions(names = ['علي', 'منى', 'عمر'], shares = [0.4, 0.4, 0.2], totalCost = 500)
```

```python
calculateSacrificeContributions(names = ['خالد'], shares = [1], totalCost = 750)
```

```python
calculateSacrificeContributions(names = ['ليلى', 'هند'], shares = [0.3, 0.7], totalCost = 300)
```

## Computational Complexity

### Time & Space Complexity: $O(n)$
