# Problem Description

<div dir="rtl" style="text-align: right;">
قُم بكتابة دالة تستقبل مصفوفة من 4 أرقام. تقوم الدالة بحساب مجموع عناصر تلك المصفوفة وإرجاع النتيجة. في حال كان أحد أرقام المصفوفة يساوي 7، فجميع الأرقام بعد ذلك الرقم لا يتم احتسابه في المجموع الكلي. 
</div>


# Solution

```python
from typing import List
def calculate_sum(lst: List[int]) -> int:
    total  = 0
    for num in lst:
        if num == 7:
            break
        total += num
    return total
```

```python
calculate_sum(lst = [1, 2, 3, 4])
```

```python
calculate_sum(lst = [0, 7, 12, 67])
```

```python
calculate_sum(lst = [23, 43, 27, 87, 67])
```

```python
calculate_sum(lst = [84, 469])
```

### Computational Complexity

#### Time Complexity: $O(n)$
#### Space Complexity: $O(1)$
