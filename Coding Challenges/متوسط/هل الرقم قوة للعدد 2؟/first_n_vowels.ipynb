# Problem Description

<div dir="rtl" style="text-align: right;">
مهمتك هي كتابة دالة تقوم بتحليل الأرقام وتحديد ما إذا كانت قوة للعدد 2 أم لا. إذا كان الرقم كذلك، يجب أن تطبع رسالة مثل:

"8 is power of 2"

أما إذا لم يكن كذلك، فيجب أن تطبع رسالة مثل:

"7 is not a power of 2"

مثال :
الادخال :
8

الاخراج :
8 is power of 2
🛠 خطوات الحل:

    أدخل رقمًا صحيحًا.

    تحقق إذا كان الرقم قوة للعدد 2 باستخدام العمليات الثنائية.

    أعد طباعة الرسالة المناسبة بناءً على النتيجة.
</div>

## Solution

```python
from typing import List
def powerof2(num: int) -> str:
    return f"{num} is power of 2" if num > 0 and (num & (num-1)) == 0 else f"{num} is not a power of 2"
```

```python
powerof2(num = 8)
```

```python
powerof2(num = 16)
```

```python
powerof2(num = 3)
```

```python
powerof2(num = 64)
```
