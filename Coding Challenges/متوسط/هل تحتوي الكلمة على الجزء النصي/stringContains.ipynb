# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة تستقبل متغيرين من نوع string ثم تقوم بإرجاع true إذا كان النص الأول يحتوي على الجزء النصي المعطى في المتغير الثاني وتقوم بإرجاع false في ما عدا ذلك
</div>


# Solution

```python
from typing import List
def stringContains(firstName: str, contains: str) -> bool:
    return contains in firstName
```

```python
stringContains(firstName = 'Ahmad', contains = 'A')
```

```python
stringContains(firstName = 'Ahmad', contains = 'N')
```

```python
stringContains(firstName = 'Sara', contains = 's')
```

```python
stringContains(firstName = 'Fahad', contains = 'Fa')
```

### Computational Complexity

#### Time Complexity: $O(n*m)$
#### Space Complexity: $O(1)$
