# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل قيمة نصية من نوع string. تقوم هذه الدالة بحساب عدد الكلمات في القيمة النصية ، ثم إرجاع عدد الكلمات من نوع integer.
</div>


## Solution

### 1. Manual Looping Method

```python
from typing import List
def countWords(txt: str) -> int:
    count = 0
    word_first_letter = False

    for ch in txt:
        if ch != ' ':
            if not word_first_letter:
                count += 1
                word_first_letter = True
        else:
            word_first_letter = False
    return count
```

```python
countWords(txt = 'Plant hope in the hearts of others')
```

```python
countWords(txt = 'Nora is a very ambitious person')
```

```python
countWords(txt = 'fine, Thank you')
```

```python
countWords(txt = 'ok')
```

#### Computational Complexity

##### Time Complexity: $O(n)$
##### Space Complexity: $O(1)$

### 2. split() Method

```python
from typing import List
def countWords(txt: str) -> int:
    return len(txt.split())
    
```

```python
countWords(txt = 'Plant hope in the hearts of others')
```

```python
countWords(txt = 'Nora is a very ambitious person')
```

```python
countWords(txt = 'fine, Thank you')
```

```python
countWords(txt = 'ok')
```

#### Computational Complexity

##### Time Complexity: $O(n)$
##### Space Complexity: $O(n)$
