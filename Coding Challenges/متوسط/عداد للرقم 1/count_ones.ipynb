قم بكتابة دالة function تستقبل قيمة نصية من نوع string ، تقوم الدالة بالتحقق هل القيمة النصية تمثل عملية رياضية أم لا ثم قم بارجاع النتيجة من نوع boolean

## Solution

### 1. Brian Kernighan's Algorithm 

```python
from typing import List
def count_ones(num: int) -> int:
    count = 0
    while (num):
        num &= num-1
        count += 1
    return count
```

```python
count_ones(num = 100)
```

```python
count_ones(num = 65)
```

```python
count_ones(num = 888)
```

```python
count_ones(num = 0)
```

#### Computational Complexity

##### Time Complexity: $O(k)$
##### Space Complexity: $O(1)$

### 2. bin().count() Method

```python
from typing import List
def count_ones(num: int) -> int:
    return bin(num)[2:].count('1')
```

```python
count_ones(num = 100)
```

```python
count_ones(num = 65)
```

```python
count_ones(num = 888)
```

```python
count_ones(num = 0)
```

#### Computational Complexity

##### Time & Space Complexity: $O(log(n))$
