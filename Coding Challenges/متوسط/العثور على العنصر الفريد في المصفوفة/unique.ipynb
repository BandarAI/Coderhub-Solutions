# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل مصفوفة من نوع integer تقوم الدالةـ function بالبحث عن الأرقام الفريدة في المصفوفة وثم تقوم بإرجاعها داخل مصفوفة جديدة.
</div>


## Solution

```python
from typing import List
def unique(arr: List[int]) -> List[int]:
    freq = {}

    for num in arr:
        freq[num] = freq.get(num, 0) + 1
    
    return [num for num in arr if freq[num] == 1]
```

```python
unique(arr = [1, 1, 1, 2, 1, 1])
```

```python
unique(arr = [3, -4, 3, 3, 3])
```

```python
unique(arr = [2, 4, -2])
```

```python
unique(arr = [2, 3, 2, 6, 2])
```

### Computational Complexity

#### Time & Space Complexity: $O(n)$
