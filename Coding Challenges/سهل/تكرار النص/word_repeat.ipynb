# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل قيمتين، قيمة نصية string وعدد صحيح موجب integer. تقوم الدالة بتكرار النص المدخل بحسب العدد الذي تم تمريره.
</div>


# Solution

```python
from typing import List
def word_repeat(word: str, n: int) -> str:
    return " ".join([word] * n)
```

```python
word_repeat(word = 'Hi', n = 2)
```

```python
word_repeat(word = 'Developers', n = 3)
```

```python
word_repeat(word = 'Red', n = 1)
```

```python
word_repeat(word = 'Great', n = 5)
```

# Computational Complexity

## Time and Space Complexity: $O(n)$
