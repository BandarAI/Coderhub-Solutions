# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة function تستقبل قيمة نصية من نوع string تقوم الـدالة function باستبدال the الى a أو an حسب القاعدة الإنجليزية ثم تقوم بإرجاع النتيجة من نوع string
</div>


## Solution

```python
def replaceThe(txt: str) -> str:
    words = txt.split()

    for i in range(len(words) - 1):
        if words[i] == "the":
            if words[i + 1][0].lower() in "aeiou":
                words[i] = "an"
            else:
                words[i] = "a"

    return " ".join(words)
```

```python
replaceThe(txt = 'the apple is on the table')
```

```python
replaceThe(txt = 'Here is the plan')
```

```python
replaceThe(txt = 'I ate the apple')
```

```python
replaceThe(txt = 'I go to the mall')
```

```python
replaceThe(txt = 'I want the orange')
```

### Computational complexity

#### Time & space Complexity $O(n)$
