# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة تستقبل متغيرين من نوع string. الدالة يجب أن تُرجع yes إذا كان من الممكن تشكيل النص الموجود في المتغير target من الأحرف الموجودة في source، دون الأخذ في الاعتبار ترتيب الأحرف أو الفرق بين الحروف الصغيرة والكبيرة. وترجع no في غير ذلك.
</div>


# Solution

```python
from typing import List
def canForm(source: str, target: str) -> str:
    source = source.lower()
    target = target.lower()

    count = {}

    for char in source:
        count[char] = count.get(char,0)+1
    
    for char in target:
        if char not in count or count[char] == 0:
            return 'no'
        count[char] -= 1
    return 'yes'
```

```python
canForm(source = 'Hubcoders', target = 'coderhub')
```

```python
canForm(source = 'SATR', target = 'satr')
```

```python
canForm(source = 'programmer', target = 'programming')
```

```python
canForm(source = 'worldhello', target = 'hello')
```
