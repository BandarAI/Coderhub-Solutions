# Solution

```python
from typing import List
def get_saudi_traditional_dance(region: str) -> str:
    region_dict = {
        'الشمالية' : 'الدحة',
        'الوسطى' : 'العرضة النجدية',
        'الجنوبية' : 'الخطوة الجنوبية',
        'الشرقية' : 'الفجري',
    }

    return region_dict.get(region, "لعبة غير معروفة" )
```

```python
get_saudi_traditional_dance(region = 'الشمالية')
```

```python
get_saudi_traditional_dance(region = 'الوسطى')
```

```python
get_saudi_traditional_dance(region = 'الجنوبية')
```

```python
get_saudi_traditional_dance(region = 'الشرقية')
```
