# Problem Description

<div dir="rtl" style="text-align: right;">
افترض أن لديك المبالغ التالية:

    اليوم 1: 50.0

    اليوم 2: 25.0

    اليوم 3: 30.0

    اليوم 4: 20.0

والهدف هو 100.0.

الحساب:

    اليوم 1: مجموع التبرعات = 50.0

    اليوم 2: مجموع التبرعات = 50.0 + 25.0 = 75.0

    اليوم 3: مجموع التبرعات = 75.0 + 30.0 = 105.0

بما أن المجموع وصل إلى 105.0 في اليوم الثالث، وهو أكبر من الهدف 100.0، فإن الدالة تُرجع القيمة 3.
</div>


## Solution

```python
from typing import List
def findDonationTargetDay(donations: List[float], target: float) -> int:
    total = 0
    for day in range(len(donations)):
        total += donations[day]

        if total >= target:
            return day+1
    
    return -1
```

```python
findDonationTargetDay(donations = [50, 25, 30, 20], target = 100)
```

```python
findDonationTargetDay(donations = [10, 15, 20], target = 50)
```

```python
findDonationTargetDay(donations = [20, 20, 20, 20, 20], target = 100)
```

```python
findDonationTargetDay(donations = [100], target = 100)
```

### Computational complexity

#### Time Complexity: $O(n)$
#### Space Complexity: $O(1)$
