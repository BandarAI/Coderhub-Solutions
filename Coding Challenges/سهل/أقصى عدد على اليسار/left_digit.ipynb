# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل قيمة نصية من نوع string ، تقوم الـدالة function باستخراج أول أقصى رقم على اليسار ، ثم تقوم الدالة بارجاع النتيجة من نوع integer
</div>


## Solution

```python
from typing import List
def left_digit(strParam: str) -> int:
    for char in strParam:
        if char.isnumeric():
            return int(char)
    return None
```

```python
left_digit(strParam = 'Welcome2back')
```

```python
left_digit(strParam = '5')
```

```python
left_digit(strParam = '9753')
```

```python
left_digit(strParam = 'nora30@gmail.com')
```

### Computational complexity

#### Time Complexity: $O(n)$
#### Space Complexity: $O(1)$
