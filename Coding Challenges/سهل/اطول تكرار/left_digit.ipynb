# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل نص string ومهمتك هي العثور على أطول تكرار للحروف في النص .
</div>


## Solution

```python
from typing import List
def repetitions(s: str) -> int:
    max_count = 1
    current_count = 0
    prev_char = None

    for char in s:
        if char == prev_char:
            current_count += 1
        else:
            current_count = 1
            prev_char = char
        
        max_count = max(max_count, current_count)

    return max_count
```

```python
repetitions(s = 'AAAAAAAAAA')
```

```python
repetitions(s = 'AAAABBBCCD')
```

```python
repetitions(s = 'AMMMDDCCCMD')
```

```python
repetitions(s = 'MZSDDDD')
```

### Computational complexity

#### Time Complexity: $O(n)$
#### Space Complexity: $O(1)$
