# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل عدد صحيح integer. تقوم الدالة بإرجاع الأعداد بشكل تنازلي من العدد المُدخل إلى القيمة صفر 0 كقيمة نصية string مفصول بين كل عنصر والآخر بمسافة.
</div>


# Solution

```python
from typing import List
def countDown(number: int) -> str:
    return ' '.join(str(num) for num in range(number, -1, -1))
```

```python
countDown(number = 10)
```

```python
countDown(number = 9)
```

```python
countDown(number = 5)
```

```python
countDown(number = 2)
```
