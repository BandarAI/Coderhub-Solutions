# Problem Description

<div dir="rtl" style="text-align: right;">
عليك أن تبدأ بدعم الحالات التي تحتاج الأقل أولًا، حتى تستطيع أن تصل لعدد أكبر قبل أن تنفد الميزانية.

المثال الإرشادي:
إذا كان لديك الطلبات [5, 3, 4, 2] والميزانية 10، ابدأ من الأصغر:

    تموّل 2 → المتبقي 8 (المموّل: 1)

    تموّل 3 → المتبقي 5 (المموّل: 2)

    تموّل 4 → المتبقي 1 (المموّل: 3)
    هكذا، وصلت الفرح إلى 3 حالات كاملة
</div>


## Solution

```python
from typing import List

def maxFundedCases(needs: List[int], budget: int) -> int:
    count = 0
    for num in sorted(needs):
        if budget >= num:
            budget -= num
            count += 1
        else:
            break
    return count
```

```python
maxFundedCases(needs = [5, 3, 4, 2], budget = 10)
```

```python
maxFundedCases(needs = [10, 10, 10], budget = 5)
```

```python
maxFundedCases(needs = [1, 1, 1, 1, 1], budget = 3)
```

```python
maxFundedCases(needs = [], budget = 100)
```

### Computational complexity

#### Time Complexity: $O(nlogn)$
#### Space Complexity: $O(n)$
