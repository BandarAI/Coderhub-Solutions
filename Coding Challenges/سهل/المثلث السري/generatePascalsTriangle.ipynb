# Problem Description

<div dir="rtl" style="text-align: right;">
مهمتك هي إنشاء مثلث باسكال بحيث:

    يتم إعطاء عدد numRows يمثل عدد الصفوف المطلوبة.

    الصف الأول يحتوي على العنصر [1].

    كل صف جديد يعتمد على السابق، بحيث يكون:

        العنصر الأول والأخير دائمًا 1.

        كل عنصر آخر هو مجموع العنصرين اللذين فوقه مباشرة.

    يتم إخراج الأرقام بتسلسل يظهر هيكل المثلث بشكل خطي.

مثال :

إذا كان numRows = 5، فإن الناتج يكون:

[

1,

1, 1,

1, 2, 1,

1, 3, 3, 1,

1, 4, 6, 4, 1

]
</div>


# Solution

```python
from typing import List
def generatePascalsTriangle(numRows: int) -> List[int]:
    res = []
    row = [1]

    for i in range(numRows):
        res.extend(row)
        row = [1] + [row[j] + row[j + 1] for j in range(len(row) - 1)] + [1]
    return res
```

```python
generatePascalsTriangle(numRows = 5)
```

```python
generatePascalsTriangle(numRows = 1)
```

```python
generatePascalsTriangle(numRows = 2)
```

```python
generatePascalsTriangle(numRows = 3)
```

## Time Complexity: $O(n)$
## Space Complexity: $O(1)$
