# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل قيمة نصية من نوع string عبارة عن على عدد عشري أو صحيح. تقوم الدالة بحساب عدد الأرقام التي بعد الفاصلة وإرجاع النتيجة على شكل قيمة من نوع integer.
</div>


# Solution

```python
from typing import List
def Decimal_places(num: str) -> int:
    if '.' in num:
        return len(num.split('.')[1])
    return 0
```

```python
Decimal_places(num = '3.967')
```

```python
Decimal_places(num = '2.9')
```

```python
Decimal_places(num = '200')
```

```python
Decimal_places(num = '5.19054')
```

## Time & Space Complexity: $O(n)$
