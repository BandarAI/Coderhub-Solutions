# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل مصفوفة array من نوع integer. تقوم الدالة بجمع أدنى رقمين في المصفوفة وإرجاع النتيجة.
</div>


## Solution

```python
from typing import List
def sum_two_smallest_nums(arr: List[int]) -> int:
    first_min = min(arr)
    arr.remove(first_min)
    second_min = min(arr)
    return first_min + second_min
```

```python
sum_two_smallest_nums(arr = [2, 5, 6, 7, 3])
```

```python
sum_two_smallest_nums(arr = [4, 3, 60, 9])
```

```python
sum_two_smallest_nums(arr = [50, 1, 1, 4])
```

```python
sum_two_smallest_nums(arr = [7, 8, 60, 100])
```

### Computational complexity

#### Time Complexity $O(n)$
#### Space Complexity $O(1)$
