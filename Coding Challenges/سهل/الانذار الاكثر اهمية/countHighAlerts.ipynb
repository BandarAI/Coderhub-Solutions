# Problem Description

<div dir="rtl" style="text-align: right;">
مهمتك هي كتابة برنامج يحسب عدد الإنذارات التي تكون درجة شدّتها أكبر من أو تساوي قيمة العتبة threshold، لتتمكن بسرعة من معرفة حجم الإنذارات الحرجة قبل اتخاذ القرار المناسب.
</div>


## Solution

```python
from typing import List
def countHighAlerts(alerts: List[int], threshold: int) -> int:
    return sum(1 for alert in alerts if alert >= threshold)
```

```python
countHighAlerts(alerts = [2, 5, 7, 1, 9], threshold = 5)
```

```python
countHighAlerts(alerts = [1, 2, 3], threshold = 0)
```

```python
countHighAlerts(alerts = [4, 4, 4, 4], threshold = 4)
```

```python
countHighAlerts(alerts = [], threshold = 10)
```

### Computational complexity

#### Time Complexity $O(n)$
#### space Complexity $O(1)$
