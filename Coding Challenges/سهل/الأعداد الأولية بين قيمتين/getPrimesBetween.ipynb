قم بكتابة الدالة function تستقبل عددين، كلاهما عدد صحيح، ثم تقوم الدالة بإرجاع مصفوفة تحتوي على جميع الأرقام الأولية بين هذين الرقمين. 
</div>


## Solution

### 1. Sieve of Eratosthenes

```python
from typing import List
def getPrimesBetween(a: int, b: int) -> List[int]:
    is_prime = [True] * (b + 1)
    is_prime[0] = is_prime[1] = False

    for n in range(2, int(b ** 0.5) + 1):
        if not is_prime[n]:
            continue
        for m in range(n + n, b + 1, n):
            is_prime[m] = False

    primes = []
    for i in range(len(is_prime)):
        if is_prime[i] and i >= a:
            primes.append(i)

    return primes
```

```python
getPrimesBetween(a=1, b=10)
```

```python
getPrimesBetween(a=20, b=50)
```

```python
getPrimesBetween(a=90, b=100)
```

```python
getPrimesBetween(a=53, b=67)
```

#### Computational complexity

##### Time Complexity $O(nlog logn)$
##### Space Complexity $O(n)$
