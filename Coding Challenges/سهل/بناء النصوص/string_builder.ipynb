# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل قيمة نصية من نوع string. تقوم بالدالة ببناء وإرجاع قيمة نصية بناءًا على القيمة المدخلة. عن طريق تكرار الاحرف بعدد معين تحدده القيمة المدخلة. على سبيل المثال:

4a يعني أنه يجب تكرار الحرف "a" أربع مرات. النتيجة المتوقعة هي aaaa. تقوم الدالة بإنشاء النص المطلوب باتباع تلك التعليمات، باستخدام مبدأ Last-In, FirstOut
</div>


## Solution

```python
from typing import List
def string_builder(expression: str) -> str:
    stack = []
    current_num = 0
    current_str = ''

    for char in expression:
        if char.isdigit():
            current_num = current_num * 10 + int(char)
        elif char == '[':
            stack.append((current_str, current_num))
            current_str = ''
            current_num = 0
        elif char == ']':
            last_str, num = stack.pop()
            current_str = last_str + current_str * num
        else:
            current_str += char
    return current_str
```

```python
string_builder(expression = '3[a]2[bc]')
```

```python
string_builder(expression = '3[a2[c]]')
```

```python
string_builder(expression = '2[abc]3[cd]ef')
```

```python
string_builder(expression = 'abc3[cd]xyz')
```

### Computational complexity

#### Time Complexity: $O(n)$
#### Space Complexity: $O(1)$
