# Problem Description

<div dir="rtl" style="text-align: right;">
مطلوب منك إكمال دالة تستقبل عددًا صحيحًا فرديًا (n)، وهو يمثل الفرق بين مربعين مثاليين متتاليين (أي عددين صحيحين متتاليين مرفوعين للقوة الثانية).
مهمتك هي إيجاد هذين العددين وإرجاع مربعيهما على شكل نص بالتنسيق التالي:
"المربع الأكبر - المربع الأصغر".

مثال :
إذا كان الإدخال n = 9:

    الفرق بين المربعين المتتاليين 25 و 16.

    الناتج: "25-16".

    إذا كان الإدخال n = 5:

        الفرق بين المربعين المتتاليين 9 و 4.

        الناتج: "9-4".

    إذا كان الإدخال n = 7:

        الفرق بين المربعين المتتاليين 16 و 9.

        الناتج: "16-9".
</div>


## Solution

```python
from typing import List
def findSquares(n: int) -> str:
    a = (n + 1) // 2
    return f"{a*a}-{(a-1)*(a-1)}"
```

```python
findSquares(n = 9)
```

```python
findSquares(n = 5)
```

```python
findSquares(n = 7)
```

```python
findSquares(n = 1)
```

### Computational complexity

#### Time & Space Complexity $O(1)$
