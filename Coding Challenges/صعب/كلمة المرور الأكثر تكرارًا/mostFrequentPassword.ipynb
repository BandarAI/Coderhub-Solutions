# Problem Description

<div dir="rtl" style="text-align: right;">
مهمتك: كتابة دالة تجد أكثر كلمة مرور مكرّرة. إذا كان هناك أكثر من كلمة مرور بنفس العدد الأعلى، نعيد الأصغر ترتيبًا أبجديًا (lexicographically).

مثال عملي:
passwords = ["1234","admin","1234","root","admin","1234"]

    "1234" تكررت 3 مرات

    "admin" تكررت مرتين

    "root" مرة واحدة

النتيجة: "1234"
</div>


## Solution

```python
from typing import List
def mostFrequentPassword(passwords: List[str]) -> str:
    dict = {}
    best_password = ""
    max_count = 0

    for password in passwords:
        if password in dict:
            dict[password] += 1
        else:
            dict[password] = 1

    for password in dict:
        if dict[password] > max_count:
            max_count = dict[password]
            best_password = password
        elif dict[password] == max_count:
            if password < best_password:
                best_password = password
    
    return best_password
```

```python
mostFrequentPassword(passwords = ['1234', 'admin', '1234', 'root', 'admin', '1234'])
```

```python
mostFrequentPassword(passwords = ['qwe', 'qwe', 'asd', 'asd'])
```

```python
mostFrequentPassword(passwords = ['pass', 'pass', 'word', 'word', 'test'])
```

```python
mostFrequentPassword(passwords = ['z', 'y', 'x', 'z', 'y', 'x'])
```

### Computational Complexity

#### Time Complexity: $O(n)$
#### Space Complexity: $O(1)$
