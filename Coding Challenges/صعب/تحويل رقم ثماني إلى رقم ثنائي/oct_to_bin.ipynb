# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل متغير من نوع int يعبر عن قيمة ثمانية octal number، ثم قم بإرجاع النتيجة بعد التحويل الى قيمة ثنائية binary number من نوع string.
 </div>


## Solution

```python
from typing import List
def oct_to_bin(octal: int) -> str:
    return bin(int(str(octal), 8))[2:]
```

```python
oct_to_bin(octal = 5252)
```

```python
oct_to_bin(octal = 777)
```

```python
oct_to_bin(octal = 123)
```

```python
oct_to_bin(octal = 6523)
```

### Computational Complexity

#### Time and Space Complexity $O(n)$
