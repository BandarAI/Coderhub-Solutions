# Problem Description

<div dir="rtl" style="text-align: right;">
قم بكتابة دالة function تستقبل عدد صحيح من نوع integer ، تقوم الدالةـ function بتحويل الـعدد الصحيح integer المدخل إلى عدد ثنائي ثم تقوم بحساب عدد تكرار الرقم 1 ، ** نتائج الدالة من نوع integer
</div>


## Solution

```python
from typing import List
def first_n_vowels(phrase: str, n: int) -> str:
    vowels = ''.join(ch for ch in phrase if ch in "aeiouAEIOU")
    if n <= len(vowels):
        return vowels[:n]
    else:
        return "invalid"
    
```

```python
first_n_vowels(phrase = 'Queen', n = 3)
```

```python
first_n_vowels(phrase = 'First Time', n = 3)
```

```python
first_n_vowels(phrase = 'Apple', n = 2)
```

```python
first_n_vowels(phrase = 'Riyadh', n = 2)
```

### Computational Complexity

#### Time Complexity: $O(n)$
#### Space Complexity: $O(1)$
